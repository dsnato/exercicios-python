###Análise da Cota para Execício da Atividade dos Senadores

```python
# Importando o Pandas
import pandas as pd
```

```python
# Carregando o arquivo csv
df = pd.read_csv('/content/despesa_ceaps_2020.csv', sep=';', encoding='latin1', skiprows=1, decimal=',')

df.shape
```

```python
#Visualizando os dados
df.head()
```

```python
# Visualizando os últimos dados
df.tail(n=10)
```

```python
# Analisando tipos e quantidades
df.info()
```

```python
# Qual valor total reembolsado?
df['VALOR_REEMBOLSADO'].sum()
```

```python
# Qual foi o maior valor reembolsado?
df['VALOR_REEMBOLSADO'].max()
```

```python
# Qual foi o reembolso de R$ 120.000,00 acima?
df[df['VALOR_REEMBOLSADO'] == df['VALOR_REEMBOLSADO'].max()]
```

```python
# Vamos verificar os 5 maiores reembolsos
df.nlargest(5, 'VALOR_REEMBOLSADO')
```

```python
# Vamos verificar os 5 menores reembolsos
df.nsmallest(5, 'VALOR_REEMBOLSADO')
```

```python
# Qual valor dos pedidos de reembolsos por Senador
df.groupby('SENADOR')['VALOR_REEMBOLSADO'].sum().sort_values(ascending=False)
```

```python
#Quantos reembolsos por tipo de pessoa
df['TIPO_DESPESA'].value_counts()
```

```python
# Quantos reembolsos pedidos por Senador?
df['SENADOR'].value_counts()
```

```python
# Subconjunto dos dados
# Vamos verificar o único reembolso do senador ORIOVISTO GUIMARÃES
df[df['SENADOR'] == 'ORIOVISTO GUIMARÃES']
```

```python
# Senador Rodrigo Pacheco foi eleito presidente do Senado

# Vamos separar os reembolsos dele

df_rodrigo_pacheco = df[df['SENADOR'] == 'RODRIGO PACHECO']

df_rodrigo_pacheco.shape
```

```python
# Somatório dos reembolsos
df_rodrigo_pacheco['VALOR_REEMBOLSADO'].sum()
```

```python
# Analisando o subconjunto do Rodrigo Pacheco
df_rodrigo_pacheco.info()
```

```python
# Vamos converter o campo DATA
df_rodrigo_pacheco['DATA'] = pd.to_datetime(df_rodrigo_pacheco['DATA'], format='%d/%m/%Y')
```

```python
# Verificando se o campo DATA foi convertido
df_rodrigo_pacheco.info()
```

```python
# Visualizando os dados
df_rodrigo_pacheco.head()
```

```python
# Plotando o gráfico de barras dos gastos por mês
df_rodrigo_pacheco.resample('M').sum()['VALOR_REEMBOLSADO'].plot(kind='bar')
```
