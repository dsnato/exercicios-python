```python
# É possível obter dados de apenas 1 coluna, ou um conjunto de colunas
# Operação de Slicing
df['sigla_uf_unidade_ensino']
```

```python
# Vamos contar quantas linhas (escolas) existem
df['sigla_uf_unidade_ensino'].value_counts()
```
